```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fb=["Facebook",2449,True,2006]
tt=["Twitter",339,False,2006]
ig=["Instagram",1000,True,2010]
yt=["YouTube",2000,False,2005]
lkn=["LinkeIn",663,False,2003]
ws=["WhatsApp",1600,True,2009]
```

```python
lista_rrss=[fb,tt,ig,yt,lkn,ws]
```

```python
df_rrss=pd.DataFrame(lista_rrss, columns = ["Nombre", "Cantidad","ES_FB", "Año"])
df_rrss
```

```python
x=[1,2,3,4,5]
y=[1,8,27,64,125]

plt.plot(x,y)
plt.show()
```

```python
#gráfico de líneas utilizando matplotlib
plt.plot(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.show()
```

```python
#gráfico de puntos utilizando matplotlib
plt.scatter(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.show()
```

```python
#cómo hacer lo mismo con métodos de un dataframe de pandas

df_rrss.plot(kind="scatter", x="Nombre", y="Cantidad")
plt.show()
```

```python
#gráfico de barras con matplotlib
plt.bar(df_rrss["Nombre"], df_rrss["Cantidad"])
plt.show()
```

```python
#gráfico de barras ordenado
df_rrss_sort=df_rrss.sort_values("Cantidad", ascending=False)
plt.bar(df_rrss_sort["Nombre"], df_rrss_sort["Cantidad"])
plt.show()
```

```python
#gráfico de barras ordenado y con color
plt.bar(df_rrss_sort["Nombre"], df_rrss_sort["Cantidad"],color=["b","r","g", "m","k", "c"])
plt.show()
```

```python
#hacer gráficas pastel
#gráfico de pie utilizando matplotlib
plt.pie(df_rrss["Cantidad"], labels=df_rrss["Nombre"])
plt.show()
```

```python
lista_col_rrss=["#3b5998","#FF0000","#25d366","#8a3ab9","#0e76a8", "#00acee"]
```

```python
plt.pie(df_rrss_sort["Cantidad"], labels=df_rrss_sort["Nombre"], colors=lista_col_rrss)
plt.show()
```
