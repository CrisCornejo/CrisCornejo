## Ejercicio 1.5. Calculadora
* Pedir al usuario dos números válidos, si no son válidos seguir pidiendo valores.
* Efectuar con ellos la suma, resta, multiplicación, división, división entera, resto, y potencias.
* Utilice control de errores para evitar el error de división entre cero.

````python
from decimal import DivisionByZero

def sumar (num1, num2):
    return (num1+num2)
def restar (num1, num2):
    return (num1-num2)
def multiplicar (num1, num2):
    return (num1*num2)
def dividir (dividendo, divisor):
    return (dividendo/divisor)
def dividir_num_enteros (dividendo, divisor):
    return (dividendo//divisor)
def resto (dividendo,divisor):
    return (dividendo%divisor)
def potencia (numero, potencia):
    return (numero**potencia)

def divide(num1,num2):
    try:
        return num1/num2
    except ZeroDivisionError:
        print ("No se puede dividir entre 0")
        return "Operación errónea"

op1=(int(input("Introduce el primer número: ")))
op2=(int(input("Introduce el segundo número: ")))
operacion=input("Introduce la operación a realizar (suma, resta, multiplica, divide, división números enteros, resto o potencia): ")

if operacion == "suma":
    print(sumar(op1, op2))
elif operacion == "resta":
    print (restar(op1, op2))
elif operacion == "multiplica":
    print(multiplicar(op1, op2))
elif operacion == "divide":
    print(dividir(op1, op2))
elif operacion == "división números enteros":
    print(dividir_num_enteros(op1,op2))
elif operacion == "resto":
    print (resto(op1, op2))
elif operacion == "potencia":
    print (potencia(op1,op2))
else:
    print ("Operación no contemplada")

print ("Operación ejecutada. Continuación de ejecución del programa")
````

## Ejercicio 1.13. Números aleatorios con y sin repetición

* Generar una tercera lista con todos los elementos que se repitan en ellas
* En esta tercera lista no debe repetirse ningún elemento.

````python
import random

def unico (x,L):
    esUnico=True
    for i in range (len (L)):
        if x ==L[i]:
            esUnico=False
            break
    return esUnico
L=[]
j=0
while j<10:
    x=random.randint (1,20)
    if unico (x, L):
        L.append(x)
        j+=1
print(L)
L.sort()
print (L)

repetidos={}
for n in L:
    if n in repetidos:
        repetidos[n]+=1
    else:
        repetidos[n]=0
print(repetidos)
print (len(repetidos))
````

## Ejercicio 1.15. Función separa

 - por otro, los impares

````python
import random

numeros_aleatorios=[random.randint(-20,20) for i in range(15)]
print(numeros_aleatorios)
pares=[]
impares=[]
for num in numeros_aleatorios:
    if(num % 2) == 0:
        pares.append(num)
    else:
        impares.append(num)



sin_rep_pares=set(pares)
lista_pares=sorted(sin_rep_pares)

sin_rep_impares=set(impares)
lista_impares=sorted(sin_rep_impares)

print(lista_pares)
print(lista_impares)


suma_pares=sum(lista_pares)
suma_impares=sum(lista_impares)

if suma_pares>suma_impares:
    print ("La suma de la lista de pares es mayor")
else:
    print("La suma de la lista de impares es mayor.")
````

* Mostrar en pantalla también el modo de llegar a ese resultado.
* Por ejemplo, si el binario es 101011, entonces se ha de imprimir el proceso que es:
 - 1\*2^5 + 0\*2^4 + 1\*2^3 + 0\*2^2 + 1\*2^1 + 1\*2^0 = 43

![Binario a decimal](https://github.com/financieras/pyCourse/blob/main/jupyter/img/binario_a_decimal.jpg?raw=1)

````python
numero_binario=(input("Introduzca un número binario: "))
print(f"Partimos del binario: {numero_binario}")

numero_decimal=0

for posicion, digito_string in enumerate (numero_binario[::-1]):
    numero_decimal+=int(digito_string)* 2 **posicion

print(f"El número decimal equivalente es {numero_decimal}")
````
